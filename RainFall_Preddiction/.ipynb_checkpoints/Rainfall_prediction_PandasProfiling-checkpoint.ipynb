## Missing values Imputation

```python
def randomsampleimputation(df,column):
    random_sample=df[column].dropna().sample(df[column].isna().sum(),random_state=120)
    random_sample.index=df[df[column].isnull()].index
    df.loc[df[column].isnull(),column]=random_sample
```

```python
randomsampleimputation(df, "Cloud9am")
randomsampleimputation(df, "Cloud3pm")
randomsampleimputation(df, "Evaporation")
randomsampleimputation(df, "Sunshine")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(["RainTomorrow","Date"],axis=1)
```

```python
y=df["RainTomorrow"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=120)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
pip install imblearn --quiet
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
sm=SMOTE(random_state=0)
x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_res)))
```

```python
sns.countplot(df["RainTomorrow"])
```

```python
sns.countplot(y_train_res)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
params =  {'penalty': ['l2'], 'C': range(1,5), 'solver': ["liblinear"],
            }



modelCV = GridSearchCV(LogisticRegression(max_iter=1000,n_jobs=1), params, cv=5, scoring='roc_auc')


modelCV.fit(x_train_res, y_train_res)
```

```python
modelCV.best_score_
```

```python
modelCV.best_params_
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn import metrics
```

```python
y_pred = modelCV.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
metrics.plot_roc_curve(modelCV, x_test, y_test)
metrics.roc_auc_score(y_test, y_pred, average=None) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train_res, y_train_res)
```

```python
y_pred = knn.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
metrics.plot_roc_curve(knn, x_test, y_test)
metrics.roc_auc_score(y_test, y_pred, average=None) 
```
